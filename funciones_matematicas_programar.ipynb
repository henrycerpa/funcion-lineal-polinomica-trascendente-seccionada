# Funciones algebraicas

En matemáticas, una función algebraica es una función que satisface una ecuación polinómica cuyos coeficientes son a su vez polinomios o monomios. 

### Función lineal

una función lineal es una función polinómica de primer grado, es decir, una función de una variable. Normalmente esta variable se denota con $$x$$
Que puede ser escrita como la suma de términos de la forma $$ax^{n}$$
donde $a$ es un número real y $n$ es un número natural.

Donde $n$ $\in (0,1)$; es decir, $n$ solo puede ser 0 o 1. Se le llama lineal dado que su representación en el plano cartesiano es una línea recta.

Esta función se puede escribir como: $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$ y son constantes reales. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. 


```
resolucion = 100                            # Puntos o resolucion que deseamos en la grafica
b = 5                                       # Definimos cualquier valor que deseamos calcular en b
m = 10                                      # Definimos cualquier valor que deseamos calcular en m

def f(x):                                   # Aqui definimos la funcion lineal
  return m*x+b                              # Esta seria la formula de la funcion

x = np.linspace(-10,10, num=resolucion)     # np.linspace genera un array NumPy formado por n números espaciados entre dos dados a partir de la resolucion
y = f(x)                                    # Definimos a Y para llamar la funcion lineal ubicada en def

fig, ax = plt.subplots()                    # Utilizamos el paquete matplotlib, ax definimos las ejes, utilizando plt.subplots para tener los ejes separados
ax.plot(x,y)                                # Le decimos al algoritmo donde salen los valores
ax.grid()                                   # Le agregamos una cuadricula
```

### Función polinómica

Una función polinómica es una relación que para cada valor de la entrada proporciona un valor que se calcula con un polinomio.

Formalmente, a cada variable $x$ le asigna un único valor, resultante de sustituirlo en el polinomio asociado a la función:

\begin{array}{rccl}P:&\mathbb {R} &\longrightarrow {}&\mathbb {R} \\&x&\mapsto &y=P(x)\end{array}

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una función que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 4x^{8} - x^{4} + 2x^{2} + 2$$

que es un polinomio de grado 8.


```
def f(x):                                            # Aqui definimos la funcion polinomica
  return (4*x**8) - (x**4) + (2*x**2) + 2             # Esta seria la formula de la funcion

y = f(x)                                             # Definimos a Y para llamar la funcion polinomica ubicada en def

plt.plot(x,y)                                        # Le decimos al algoritmo donde salen los valores      
plt.grid()                                           # Le agregamos una cuadricula
```

### Función potencia

Son un caso especial de la función polinómica. La función potencia está definida por la expresión general $f(x) = ax^n$, el exponente $n$ puede ser un número entero o real (positivo o negativo), el factor $a$ es real.

Esta función tiene la siguiente forma:

$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$


```
def f(x):                 # Aqui definimos la funcion potencia
  return x**2             # Esta seria la formula de la funcion

y = f(x)                  # Definimos a Y para llamar la funcion potencia ubicada en def

plt.plot(x,y)             # Le decimos al algoritmo donde salen los valores      
plt.grid()                # Le agregamos una cuadricula
```

# Funciones trascendentes

Son funciones que no pueden ser expresadas con polinomios. 

### Funciones trigonométricas

Algunos ejemplos son las funciones $cos(x)$, $sen(x)$ y $tan(x)$

```
def f(x):
  return np.cos(x)

y = f(x)

plt.plot(x,y)
```

### Función exponencial

Tienen la forma de $$f(x)=a^x$$ donde la base $a$ es una constante positiva. Un gran ejemplo de una función exponencial es usando la base como el número de euler:

$$f(x)=e^x$$

```
def f(x):
  return np.exp(x)

y=f(x)

plt.plot(x,y)
```

### Función logaritmo

El logaritmo está definido por la **relación**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que está elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$





```
def f(x):
  return np.log2(x)

x = np.linspace(0.001,256, num=1000)
plt.plot(x,f(x))
```

# Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    

N=1000

x = np.linspace(-10,10, num=N)

y = H(x)

plt.plot(x,y)
```
